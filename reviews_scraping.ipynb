## **Scraping Airline Reviews**

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re

headers = {
    "user-agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36",
    "Accept-Encoding": "gzip, deflate, br",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7",
    "DNT": "1", "Connection": "keep-alive", "Upgrade-Insecure-Requests": "1"}

# URL for airline reviews
url = "https://www.airlinequality.com/review-pages/a-z-airline-reviews/"
response = requests.get(url)
content = response.content
soup = BeautifulSoup(content, "html.parser")

# Extracting alphabets from the page
ul = soup.find("ul", class_="tabs")
alphabets = [s.find("a").text for s in ul]
print(alphabets)

# Get all airline names
href_pattern = re.compile(r'/airline-reviews/')
airline_links = soup.find_all("a", href=href_pattern)

# Extracting airline names from links
airlines = [link.text for link in airline_links]

# Cleaning and formatting airline names
airline_names = [name.lower().replace(" ", "-") for name in airlines[:560]]
airline_names = ' '.join(airline_names).replace("t'way-air", "tway-air").split()
# print(airline_names)
```

```python
import requests
from bs4 import BeautifulSoup

url_format = "https://www.airlinequality.com/airline-reviews/{0}/page/{1}/"
pageNo = 100

def get_reviews(pageName, pageNo):
    headers = {
        "user-agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36",
        "Accept-Encoding": "gzip, deflate, br",
        "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7",
        "DNT": "1", "Connection": "close", "Upgrade-Insecure-Requests": "1"
    }

    r = requests.get(url_format.format(pageName, pageNo))
    content = r.content
    soup = BeautifulSoup(content, "html.parser")

    all_reviews = []
    
    for rev in soup.find_all("article", itemprop='review'):
        data_dct = {}
        
        rating = rev.find("span", itemprop="ratingValue")
        subject = rev.find("h2", class_="text_header")
        name = rev.find("h3", class_="text_sub_header userStatusWrapper")
        trip_status = rev.find("em")
        review = rev.find("div", class_="text_content")
        airline = soup.find("h1", itemprop="name")
        
        table = rev.find("table", class_="review-ratings")
        
        for e in table.find_all("tr"):
            rev_head = e.find("td", class_="review-rating-header")
            rev_val = e.find("td", class_="review-value")
          
            if rev_val is not None and len(rev_val) > 0:
                data_dct[rev_head.text] = rev_val.text
                
        for i in table.find_all("tr"):
            rtns_head = i.find("td", class_="review-rating-header")
            rtns = i.find("td", class_="review-rating-stars stars")
            star_rtns = i.find_all("span", class_="star fill")
            
            if rtns is not None and len(star_rtns) > 0:
                data_dct[rtns_head.text] = star_rtns[-1].text
        
        all_col = [rating, subject, name, trip_status, review, airline]
        all_col_names = ['Rating', 'Subject', 'Name', 'Trip Status', 'Review', "Airline"]
        
        for col_name, col_value in zip(all_col_names, all_col):
            if col_value is not None:
                data_dct[col_name] = col_value.text.strip()
            else: 
                data_dct[col_name] = "unk"

        all_reviews.append(data_dct)

    return all_reviews



```

```python

# Example usage
# get_reviews("air-china", 1)
import time

max_retries = 7
retries = 0

while retries < max_retries:
    try:
        all_reviews = []
        for airline_name in airline_names:
            for page_number in range(1, pageNo + 1):
                reviews = get_reviews(airline_name, page_number)
                if reviews:
                    all_reviews.extend(reviews)
                else:
                    continue

#         break  # Break the loop if the request is successful
    except requests.exceptions.RequestException as e:
        print(f"Error: {e}")
        retries += 1
        time.sleep(1)  # Wait for a short time before retrying
```
